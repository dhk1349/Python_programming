```python
import scipy
import numpy as np
from scipy.spatial import Voronoi, voronoi_plot_2d
import matplotlib.pyplot as plt
```

```python
points = np.array([[1, 0], [0, 1],[0,0], [1/2, 1/2], [1,1/2],[0,1/2]])
#points = np.array([[1, 0], [0, 1],[1,1],[0,0], [1/2, 1/2]])
```

```python
vor = Voronoi(points)
```

```python
fig = voronoi_plot_2d(vor)
plt.show()
```
