```python
def combiner(path, filename):
    year=['2016', '2017', '2018', '2019', '2020']
    team=['HT', 'LT', 'SK', 'LG', 'NC', 'KT', 'HH', 'OB', 'SS', 'WO']
    #open files
    #path="_test_convert"
    #filename="_test_convert_bigcon_"+year[-2]+"_"

    index_front = ["G_ID", "GDAY_DS", "T_ID", "VS_T_ID", "HEADER_NO", "TB_SC",
             "WHIP", "LOB", "ERA", "FIP", "H_9", "K_9", "BB_9", "HR_9",
             "oAVG", "oSLG", "oOBP", "P_WHIP_RT", "P2_WHIP_RT", "CB_WHIP_RT",
             "AVG", "OBP", "SLG", "OPS", "RC", "GPA", "BA", "SECA", "TA", "XR", "ISO",
             "wOBA", "P_HRA_RT", "DER", "PF", "H_A", "WLS"]

    index_back = ["WHIP", "LOB", "ERA", "FIP", "H_9", "K_9", "BB_9", "HR_9",
             "oAVG", "oSLG", "oOBP", "P_WHIP_RT", "P2_WHIP_RT", "CB_WHIP_RT",
             "AVG", "OBP", "SLG", "OPS", "RC", "GPA", "BA", "SECA", "TA", "XR", "ISO",
             "wOBA", "P_HRA_RT", "DER", "PF", "H_A", "WLS"]

    index_back=["vs_"+i for i in index_back]

    #index_back[0]=index_back[0][3:]
    if (os.path.isdir(os.getcwd()+"\\combined")==False):
        os.mkdir('combined')
    dfs=[]

    df = pd.DataFrame(columns = index_front)
    for y in year:
        for i in team:
            filepath=str(y)+path+"\\"+filename+str(y)+"_"+i+".csv"
            file = pd.read_csv(filepath, index_col=None, header=0)
     
            dfs.append(file)

        df = pd.concat(dfs, axis=0, ignore_index=True)
        df=df.sort_values(by=['G_ID'], axis=0)
    
        new_df = pd.DataFrame(columns = index_front+index_back[1:])

        for idx,_row in enumerate(df.iterrows()):
            if (idx%2==0):
                eve_row=pd.DataFrame(df[idx:idx+1], columns=index_front)
                eve_row=eve_row.reset_index()
                eve_row=eve_row.drop(columns=['index'])
            else:
                odd_row=pd.DataFrame(df[idx:idx+1], columns=index_front)
                odd_row=odd_row.drop(columns=index_front[:6])
                odd_row.columns=index_back
                odd_row=odd_row.reset_index()
                odd_row=odd_row.drop(columns=['index'])        
        
                row=pd.DataFrame(pd.concat([eve_row, odd_row], axis=1))
        
                new_df=new_df.append(row, ignore_index=True)[row.columns.tolist()]
        #print(new_df)
        new_df.to_csv("combined\combined_"+str(y)+".csv")
    
```
